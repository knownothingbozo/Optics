```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
import h5py
```

```python
f1 = h5py.File('build/data/ey-000000050.h5', 'r')
f2 = h5py.File('build/data/ey-000000200.h5', 'r')
list(f1.keys())
```

```python
dset1 = f1['ey.r']
dset2 = f2['ey.r']
dset1.shape
lim = 1.25*np.max(dset1)
max(dset1[:,int(dset1.shape[1]/2)])
```

```python
fig = plt.figure(figsize=(10,4), layout='constrained')
fig.suptitle(r'Electric field')
ax1 = fig.add_subplot(1, 2, 1, title='', xlabel=r'$x$', ylabel=r'$E_y$')
ax1.set_title('t=340.50 fs')
ax1.set_ylim([-lim, lim])
ax1.set_xlim([0, 100])
ax1.plot(np.linspace(0.0, dset1.shape[0], dset1.shape[0]), dset1[:,int(dset1.shape[1]/2)])
ax1 = fig.add_subplot(1, 2, 2, title='', xlabel=r'$x$', ylabel=r'$E_y$')
ax1.set_title('t=964.50 fs')
ax1.plot(np.linspace(0.0, 1.0, dset1.shape[1]),dset1[80,:])
#ax1.legend()
plt.show;
```
